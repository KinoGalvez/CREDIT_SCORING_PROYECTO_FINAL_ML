```python
# Crear un DataFrame vacío para almacenar los resultados
resultados_por_año = pd.DataFrame()

# Iterar sobre los años
for año in range(2009, 2016):
    # Filtrar el DataFrame para el año específico
    lc_año = lc[(lc['issue_d'].dt.year == año)]
    
    # Guardar el DataFrame filtrado en el DataFrame de resultados
    resultados_por_año = pd.concat([resultados_por_año, lc_año])
    
    # Mostrar la distribución de 'loan_status' para el año específico
    print(f'Loan Status Distribution for {año}:')
    print(lc_año['loan_status'].value_counts(normalize=True))
    print('\n')

# Mostrar el DataFrame completo de resultados
print('DataFrame completo con datos por año:')
print(resultados_por_año)
```

```python
# Crear un diccionario para almacenar el conteo por año
conteo_por_año = {}

# Iterar sobre los años
for año in range(2009, 2016):
    # Filtrar el DataFrame para el año específico
    lc_año = lc[(lc['issue_d'].dt.year == año)]
    
    # Almacenar el conteo en el diccionario
    conteo_por_año[año] = len(lc_año)

# Mostrar el conteo por año
for año, conteo in conteo_por_año.items():
    print(f'Número de elementos para el año {año}: {conteo}')
```
